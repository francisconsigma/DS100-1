```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the URL
url = "https://raw.githubusercontent.com/francisconsigma/DS100-1/main/public-health-expenditure-share-GDP-OWID.csv"

# Read the CSV file from the URL
data = pd.read_csv(url)

# Filter the data for the desired columns
filtered_data = data[['Code', 'Year', 'public_health_expenditure_pc_gdp']]

# Sort the data by the public_health_expenditure_pc_gdp column in descending order
sorted_data = filtered_data.sort_values(by='public_health_expenditure_pc_gdp', ascending=False)

# Define custom color palette
colors = ['steelblue', 'forestgreen', 'orange', 'purple', 'salmon']

# Plotting the horizontal bar graph
plt.figure(figsize=(12, 8))
bars = plt.barh(sorted_data['Year'], sorted_data['public_health_expenditure_pc_gdp'], color=colors)
plt.xlabel('Public Health Expenditure per Capita (% of GDP)')
plt.ylabel('Year')
plt.title('Public Health Expenditure per Capita (% of GDP) by Year')
plt.grid(True)

# Create a legend with the 'Code' values
legend_labels = sorted_data['Code'].unique()
plt.legend(bars, legend_labels, loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust the margins

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('public-health-expenditure-share-GDP-OWID.csv')

# Select the required columns
data = data[['Code', 'Year', 'public_health_expenditure_pc_gdp']]

# Pivot the data to have years as columns
data_pivot = data.pivot(index='Code', columns='Year', values='public_health_expenditure_pc_gdp')

# Sort the data by country code
data_pivot.sort_index(inplace=True)

# Plot the stairs plot
fig, ax = plt.subplots(figsize=(10, 6))
for i, country in enumerate(data_pivot.index):
    y = data_pivot.loc[country].dropna()
    x = y.index.astype(int)
    plt.step(x, y, label=country)

plt.xlabel('Year')
plt.ylabel('Public Health Expenditure (% of GDP)')
plt.title('Public Health Expenditure Over Time')

# Adjust the legend position
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1))

plt.xticks(rotation=45)

# Adjust the figure layout manually
plt.subplots_adjust(right=0.8)

plt.show()
```
